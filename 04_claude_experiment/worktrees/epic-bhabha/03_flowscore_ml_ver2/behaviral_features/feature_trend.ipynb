```python
import pandas as pd

# 1. 데이터 다시 로드
file_path = "AUS_v2_Ready_To_Train.csv"
try:
    df = pd.read_csv(file_path)
    print("✅ 데이터 로드 완료!")
except FileNotFoundError:
    print("❌ 파일을 찾을 수 없습니다. 경로를 확인해주세요.")

# 2. 기업별 레코드 수(데이터 포인트) 확인
company_counts = df['COMPANY_ID_NORM'].value_counts()

# 3. 분석 결과 도출
total_companies = len(company_counts)
multi_year_companies = len(company_counts[company_counts >= 2])
max_records = company_counts.max()

print("-" * 50)
print(f"📊 분석 대상 기업 총계: {total_companies:,}개")
print(f"📈 2개년 이상 기록 보유 기업: {multi_year_companies:,}개 ({multi_year_companies/total_companies:.1%})")
print(f"📅 한 기업당 최대 기록 수: {max_records}개")
print("-" * 50)

# 4. 연도 컬럼 존재 여부 확인
year_cols = [col for col in df.columns if 'YEAR' in col.upper() or 'DATE' in col.upper()]
if year_cols:
    print(f"🔎 발견된 시간 관련 컬럼: {year_cols}")
    for col in year_cols:
        print(f"   - {col} 범위: {df[col].min()} ~ {df[col].max()}")
else:
    print("⚠️ 시간(YEAR/DATE) 관련 컬럼이 보이지 않습니다. 추세 분석을 위해선 추가 Join이 필요할 수 있습니다.")
```
